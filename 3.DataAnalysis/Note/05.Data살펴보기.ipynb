### 자동차 연비 DataSet
- mpg : 연비
- cylinders : 실린더 수
- displacement : 배기량
- horsepower : 출력
- weight : 차중
- acceleration : 가속 능력
- model-year : 출시년도
- origin : 제조국(1:USA, 2:EU, 3:JPN)
- name : 모델명

### Pandas의 내장 그래프 도구

```python
df = pd.read_excel("../Data/남북한발전전력량.xlsx")
df
```

```python
df_sn = df.iloc[[0,5],2:]
df_sn.index = ['South','North']
df_sn
```

```python
df_sn = df.loc[[0,5],'1990':]
df_sn.index = ['South','North']
df_sn
```

```python
df_sn = df.loc[[0,5],'1990':]
df_sn.index = ['South','North']
df_sn

df_sn.columns = df_sn.columns.map(int) # 열 이름의 자료형을 정수형으로 변경
```

```python
df_sn = df.loc[[0,5],'1990':]
df_sn.index = ['South','North']
df_sn
```

```python
# 선 그래프 그리기
df_sn.T.plot()
```

```python
# 막대 그래프 그리기
tdf_sn = df_sn.T
tdf_sn.plot(kind='bar')
```

```python
# 히스토그램
tdf_sn.plot(kind = 'hist')
```
