# 난수 발생
- rand = 0부터 1사이의 균일 분포
- randn = 가우시안 표준 정규 분포
- randint : 균일 분포의 정수 난수

### Boolean Indexing

```python
names = np.array(["Charles", "Kenny", "Hayoung", "Charles", "Hayoung", "Kenny", "Kenny"])
data = np.random.randn(7, 4)
```

```python
names.shape
```

```python
data
```

```python
data.shape
```

```python
# Mask를 이용한 추출
names == "Charles"
```

```python
data[names == "Charles", :]
```

```python
data[(names == "Charles") | (names == "Kenny"), :]
```

```python
data[:]
```

```python
data[:,3]
```

```python
# data의 4번째 열 값중 0보다 작은 것의 mask만들기

data[data[:, 3] < 0,: ]
```

```python
# data의 4번째 열의 값이 0보다 작은 data들을 0으로 대체하기

data[data[:, 3] < 0, :] = 0
data
```

# reshape***

```python
# reshape가 제일 중요
arr = np.arange(1,9)
arr
```

```python
# 2차원 배열로 변경
arr1 = np.reshape(arr, (2,4))
arr1
```

```python
# 3차원 배열로 변경
arr2 = np.reshape(arr, [2,2,2])
arr2
```

```python
# 2차원 배열을 1차원으로 변경
arr3 = arr1.reshape(-1, 1)  # 데이터 개수를 정할 수 없을 때 -1 사용
arr3                        # 딥러닝에서는 1을 표현
```

```python
arr4 = arr2.reshape(-1)
arr4
```

```python
# 단위행렬 : 
arr3 == arr4
```
