### 행선택
- index 이름을 기준으로 선택할 경우에는 loc
- 정수형 위치 index를 사용할 경우에는 iloc

```python
df
```

```python
# index가 서준인 행을 선택
print(df.loc['서준'])
print(df.iloc[0])
```

```python
# 행 인덱스를 사용하여 서준과 우현 2개의 행을 선택
# loc 사용
print(df.loc[['서준','우현']])

# iloc 사용
print(df.iloc[:2])
```

# 행과 열을 이용한 데이터 선택

```python
exam_data = {'수학':[90,80,70], '영어':[98,89,95],
 '음악':[85,95,100], '체육':[100,90,90], '이름': ['서준','우현','인아'] }
df = pd.DataFrame(exam_data)
df
```

```python
df.set_index('이름', inplace=True)
df
```

```python
print(df.loc['서준',['음악','체육']])
print(df.iloc[0,2:4])
print(df.loc['서준','음악':'체육'])
print(df.loc['서준','음악':])
```

```python
df['국어'] = 80
df
```

```python
df.loc['철수'] = [10, 20, 30, 40, 50]
df
```

```python
df.loc['영희'] = df.loc['인아']
df
```

```python
# 서준의 체육점수를 80점으로 변경하기
df.loc['서준','체육'] = 80
df
```

```python
# 서준의 음악과 체육점수를 50점으로 변경하기
df.loc['서준',['음악','체육']] = 50
df
```

```python
# 서준의 음악 점수를 100 점 체육점수를 50점으로 변경하기

df.loc['서준',['음악','체육']] = 100, 50
df
```

### DataFrame 연산

```python
import seaborn as sns
```

```python
# 타이타닉 데이터 불러오기

titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# titanic에서 age와 fare부분의 컬럼만 발췌하여 df라는 dataFrame 만들기
# df = titanic.loc[:,['age','fare']]
df = pd.DataFrame(titanic[['age','fare']])
df.head()
```

```python
# 모든 컬럼을 10씩 증가하여 addition DataFrame 생성하기
addition = df + 10
addition.head()
```

```python
# addition과 df의 차를 구하여 subtraction이라는 Dataframe 만들기

subtraction = addition - df
subtraction.head()
```
