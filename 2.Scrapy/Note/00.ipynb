# 인터넷의 빅데이터
- 빅데이터는 이름 그대로 대규모의 데이터 집합을 의미.
- 데이터를 수집하는 것만으로는 어떤 의미도 없음.
- 수집한 데이터를 활용해야 가치가 생성.
### <i> 빅데이터라는 용어는 수집한 데이터를 분석해서 비즈니스에 활용하는 것까지를 의미함.

- 빅데이터를 분석한다는 것은 수많은 데이터에서 규칙성을 찾는 것을 말함.
- 데이터에 포함되어 있는 요소 중에서 중요한 것들을 추출하고 이를 분류해서 규칙성을 찾는 것이 일반적이다.

### 블로그와 SNS
- 트랜드 분석

### 인터넷 전자 상거래
- 상품 데이터 베이스
- 네이버와 다음 같은 사이트에서 다양한 곳의 상품데이터를 모아 웹 API로 제공

### 금융정보
- 환율, 주식등의 다양한 금융정보
- 각 국가의 환율, 주식, 금값 등을 실시간으로 추출 가능.
- 이러한 정보를 정기적으로 추출해서 저장해 두고 활용하면 예측등에 활용할 수 있음

### 이미지 데이터
- 플리커, 인스타그램 같은 사이트에서 이미지 서비스를 활용하여 이미지를 획득 가능
- 이미지와 함께 제공되는 태그 정보등을 활용

### 행정기관 정보
- 공개 데이터(서울 열린 데이터 광장)
