# 크롤링과 스크래핑
### requests
#### 응답코드
- 1xx : 요청을 받았고 작업 진행중.
- 2xx : 사용자의 요청이 성공적으로 수행됨.
- 3xx : 요청은 완료되었으나 리다이렉션이 필요.
- 4xx : 사용자의 요청이 잘못됨
- 5xx : 서버에 오류가 발생함

```python
# request 응답 코드
import requests

URL = "https://www.naver.com"
response = requests.get(URL)
print(response.status_code)
```

```python
print(response.text)
```

```python
# 네이버에서 정보 검색 결과 가져오기
import requests

URL = "https://search.naver.com/search.naver"
params = {'query' : '플로터'}
response = requests.get(URL, params=params)
print(response.status_code)
print(response.text)
```

```python
# 네이버에서 정보 검색 결과 가져오기
import requests

URL = "https://comic.naver.com/webtoon/detail.nhn"
params = {'titleId' : '687915', 'no': '157'}
response = requests.get(URL, params=params)
print(response.status_code)
print(response.text)
```

```python
# Json형식의 데이터 가져오기

import requests

response = requests.get("https://raw.githubusercontent.com/naver/naver-openapi-guide/draft/naver-openapi-swagger.json")
result = response.json()
print(result)
print(type(result))
```
