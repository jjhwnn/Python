# Dictionary
: {key1 : Value1, key2 : Value2, key3 : Value3,...}

```python
dic = {'name': 'petty', 'phone': '010-1111-2222', 'birth':'1231'}
```

```python
# 딕셔너리는 key값으로 데이터를 불러온다
dic['name']
```

```python
dic = {'a' : [1,2,3]}
dic ['a'][0]
```

```python
# Dictionary에 요소 추가
dic['b'] = 'bbb'
dic
```

```python
dic['name'] = 'petty'
dic
```

```python
dic['name'] = "James" # key값이 같으면 값이 대체함
dic
```

```python
del dic['b']
dic
```

```python
# 딕셔너리 관련 함수

dic = {'name': 'petty', 'phone': '010-1111-2222', 'birth':'1231'}
dic
```

```python
dic.keys() # key값 출력
```

```python
for k in dic.keys():
    print(k)
```

```python
dic.values() # value값 출력
```

```python
dic_list = list(dic.values())
dic_list[0]
```

```python
# 해당 key존재 여부 확인

print('name' in dic)
print('email' in dic)
```
