학생들 점수의 평균을 구하여 평균점수 보다 낮으면 불합격 \n
평균점수보다 같거나 높으면 합격을 위의 출려과 같이 보여 주기

```python
import numpy as np

scores = [90, 25, 67, 45, 80]
number = 0
total = 0

avg = np.mean(scores)
print("학급의 평균 점수는 %0.2f 입니다. \n"%(avg))

for score in scores:
    total += score
print("학급의 평균 점수는 %0.2f 입니다. \n"%(total/len(scores)))

for score in scores:
    print("%d의 학생은 점수가 %d이므로 합격 입니다."%(number+1, score)) \
    if score >= 60 else print("%d의 학생은 점수가 %d이어서 불합격 입니다."%(number+1, score))
```

```python
import numpy as np

scores = [90, 25, 67, 45, 80]
number = 0
avg = np.mean(scores)

for score in scores:
    if score >= avg:
        print(f"{number+1}의 학생은 점수가 %d이므로 합격 입니다."%score)
    number += 1
```

```python
총 5명의 학생이 시험을 보았는데 시험점수가 60점을 넘으면 합격이고 그렇지 않으면 불합격이다.
합격인 학생들의 번호만 새로운 list를 생성하여 출력하기
```

```python
scores = [90, 25, 67, 45, 80]
number = 0

list60 = []


for number in range(len(scores)):
    if scores[number] >= 60:
        list60.append(number+1)
    number += 1

print("합격자 번호")
print("-"*10)    
for i in list60:
    print('{0:^10}'.format(i))
```

# 문장과 단어 추출

```python
string = """나는 홍길동 입니다.
주소는 서울시 입니다.
나이는 35세 입니다."""
```

```python
string
```

```python
sents = [] # 문장 저장
words = [] # 단어 저장

# 문단 -> 문장
for sen in string.split('\n'):
    sents.append(sen)
    # 문장 -> 단어
    for word in sen.split():
        words.append(word)

print(sents)
print(words)
```

```python
# 함수에 입력되는 인수가 몇개일지 모를경우 *

def sumManyF(*inputValue):
    sum = 0
    for i in inputValue:
        sum += i
    return sum
```

```python
# 여러개의 숫자를 받아 최대값과 최소값의 차이를 구하는 함수 만들기
import numpy as np

def sumManyF2(*inputValue):

    val = max(inputValue) - min(inputValue)
    
    return val

sumManyF2(1,3,5,7,9)
```

```python
def sumMulF(sic, *num):
    # sum = 0
    # mul = 1
    if sic == 'sum':
        # for i in num:
        #     sum += i
        # return sum
        
        # 누적합
        return np.sum(num)
    elif sic == 'mul':
        # for i in num:
        #     mul *= i
        # return mul

        # 누적곱
        return np.prod(num)
    else :
        return "What are you talking about?"
```

```python
result = sumMulF('sum', 1,2,3,4,5)
print(result)
```

```python
result = sumMulF('mul', 1,2,3,4,5)
print(result)
```

```python
result = sumMulF('add', 1,2,3,4,5)
print(result)
```

```python
# 파이선은 결과값을 튜플로 2개 이상 받을 수 있다.
def sum_mul(a, b):
    return a + b, a * b
    
```

```python
sum_mul(3,4)
```

```python
return_sum, return_mul = sum_mul(3,4)
print(return_sum)
print(return_mul)
```

```python
# 함수의 초기값 설정하기
def sum_mul(a, b=10):
    return (a + b, a * b)
    
```

```python
# b = 10 <- 가장 많이 사용되는 숫자
# 초기값이 설정되어 있으면, 파라미터로 값을 안넘겨줄때 초기값 사용
# 파라미터로 넘어올 시에는 해당 파라미터 값으로 대체됨
sum_mul(10, 20)
```

# 람다(lambda) 함수
- 보통 함수를 한줄로 간결하게 만들때 사용
- return 명령어가 필요 없다.

```python
def addFunction(a, b):
    return a + b
```

```python
addFunction(10, 20)
```

```python
addLambda = lambda a, b: a + b
addLambda(10, 20)
```
