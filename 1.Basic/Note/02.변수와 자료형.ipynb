### 날짜형 변환

```python
# 날짜형 모듈 불러오기
from datetime import datetime, timedelta
```

```python
time1 = datetime(2022, 12 , 25, 0, 0, 0)
print(time1)
```

```python
# 날짜 계산하기
time2 = datetime.now()
print(time2)
print(time1 - time2)
```

```python
# 좀 더 편리하게 출력
print(time1)
print(time2)
print(time1 - time2)
print((time1 - time2).days, '일')
print((time1 - time2).seconds, '초') # 시간 끼리의 초를 계산
```

```python
# 시간으로 확인하기
(time1 - time2).days * 24 + math.floor((time1 - time2).seconds / 60 / 60 )
```

```python
# 특정 날짜로 부터 시간 계산 ( 몇일 전, 몇일 뒤)
time2 + timedelta(days=-3)
```

```python
print(time2 + timedelta(days=1, hours=-2)) # 1일뒤 2시간 전
```
