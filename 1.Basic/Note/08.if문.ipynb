```python
number = int(input("숫자를 입력하세요"))

if number >= 90:
    print("점수는", number, "이고 A학점 입니다.")
elif number >= 80:
    print("점수는", number, "이고 B학점 입니다.")
elif number >= 70:
    print("점수는", number, "이고 C학점 입니다.")
elif number >= 60:
    print("점수는", number, "이고 D학점 입니다.")
else:
    print("점수는", number, "이고 F학점 입니다.")
```

```python
# 사용자로 부터 숫자를 입력받아 5의 배수이면 제곱근을 구하고 그렇지 않으면 원래 수를 출력하기
import math

num = int(input("숫자를 입력하세요:"))

if num % 5 == 0 :
    print(f"{num}는 5의 배수입니다 {num}의 제곱근 : {math.sqrt(num)}")
else :
    print(num, "은 5의 배수가 아닙니다")
```

```python
# list의 내용을 if문으로 확인하기

pocket = ['paper','smartphone']
card = 1

if 'money' in pocket:
    print("현금으로 택시타기!")
elif card:
    print("신용카드를 사용한 택시 타기!")
else :
    print("튼튼한 다리 사용 !")

print("--- End ---")
```

```python
# 삼항 연산자
# 변수 = 참인 경우 if(조건문) else (거짓인 경우)

num = 9
result = 0

result = num * 2  if num >= 5 else num + 2
print(result)
```
