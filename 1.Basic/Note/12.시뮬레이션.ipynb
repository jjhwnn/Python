# 몬테카를로 시뮬레이션
: 자료가 없을 경우 난수를 통해서 가설 증명

```python
# 동전의 앞면과 뒷면이 나올 확률이 같다는 것을 증명
```

```python
import random
```

```python
random.randint(1, 10)
```

```python
def coinF(n):
    coinFront = coinRear = 0
    for i in range(1,n+1):
        x = random.randint(1,2)
        if x == 1:
            coinFront += 1
        else :
            coinRear += 1
    
    print("%7d번 시도시 동전의 앞면은 %6d번이고 동전의 뒷면은 %6d번 입니다."%(n, coinFront, coinRear))
```

```python
coinF(10)
coinF(100)
coinF(1000)
coinF(10000)
coinF(100000)
coinF(1000000)
```

```python
def rps(n):
    rock = paper = scissors = 0

    for i in range(1, n+1):

        x = random.randint(1,3)

        if x == 1:
            rock += 1
        elif x == 2:
            paper += 1
        else :
            scissors += 1
            
    print("%7d번 시도시 가위는 %6d번이고 바위는 %6d번이고 보는 %6d번 입니다."%(n, scissors, rock, paper))
```

```python
rps(10)
rps(100)
rps(1000)
rps(10000)
rps(100000)
rps(1000000)
```
