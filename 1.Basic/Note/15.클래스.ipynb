```python
class Services:
    name = ""
    secret = "임금님 귀는 당나귀 귀다!"
    def __init__(self, name):
        self.name = name
    def sum(self, num01, num02):
        result = num01 +num02
        print("%s님! %d + %d = %d 입니다."%(self.name, num01, num02, result))
```

```python
instance01 = Services("관우")
instance01.sum(2,3)
```

```python
class Calc:
    num1 = 0
    num2 = 0
    
    def setData(self, num1, num2):
        self.num1 = num1    # 따로 변수 선언을 안해도 self로 접근시 전역변수로 사용됨
        self.num2 = num2

    # 덧셈
    def sum(self):
        return self.num1 + self.num2

    # 곱셈
    def mul(self):
        return self.num1 * self.num2

    # 뺄셈
    def sub(self):
        return self.num1 - self.num2
    
    # 나눗셈
    def div(self):
        return self.num1 / self.num2
    
```

```python
calc01 = Calc()
calc01.setData(4, 2)
print("덧셈", calc01.sum())
print("곱셈", calc01.mul())
print("뺄셈", calc01.sub())
print("나눗셈", calc01.div())
```
