# 파일 읽고 쓰기

```python
# 파일을 쓰기 모드로 열어 출력값 저장하기

f = open("../Data/new.txt", "w")
for i in range(1, 11):
    data = "%d번째 줄 입니다.\n"%i
    f.write(data)
f.close()
```

```python
# 파일을 읽기모드로 열어 파일 내용 출력하기 : #1
f = open("../Data/new.txt", "r")
while True:
    line = f.readline() # 한줄 읽어오기
    if not line:
        break
    print(line)
f.close()
```
