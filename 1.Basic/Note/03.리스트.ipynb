# 리스트의 연산

```python
var01 = [1,2,3]
var02 = [4,5,6]
var01 + var02
```

```python
# 리스트의 덧셈 : for문으로 구성해 보기
# in 앞에는 연속된 (iterable) 한 데이터만 들어간다.

tempList = []
for i in range(len(var01)):
    print(var01[i] + var02[i])
    tempList.append(var01[i] + var02[i])    # 리스트 추가하기

```

### 리스트 관련 함수

```python
var = [1,2,3]
var.append(4)
var
```

```python
var.append([5,6])
var
```

```python
# 리스트 정수 정렬
var = [1,4,3,2]
var.sort() # 오름차순 정렬
print(var)

var.reverse() # 내림차순 정렬
print(var)
```

```python
# 리스트의 위치정보

var = [4,5,6]
var.index(5)
```

```python
# 리스트에 요소 삽입
var = [1,2,3]
var.insert(0, 4)
var
```

```python
# 리스트에서 요소 제거 (인덱스가 아닌 데이터값을 넣어야 함)
var.remove(4)
var
```

```python
var.pop() # 마지막 인덱스의 데이터를 없앤 후 출력
var
```

```python
var = [1,2]
var.extend([3,4]) # 반복적인 데이터를 삽입
var
```
