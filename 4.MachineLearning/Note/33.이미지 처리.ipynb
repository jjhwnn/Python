```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Numpy Data л¶Ҳлҹ¬мҳӨкё°
import numpy as np
fruits = np.load("../Data/fruits.npy")
fruits.shape
```

```python
# ImageлЎң м ҖмһҘн•ҳкё°
from PIL import Image
for i in range(0,300):
    image = Image.fromarray(fruits[i]).convert("RGB")
    image.save("../Data/fruits/fruits_%03d.jpg"%i, "JPEG")
```
