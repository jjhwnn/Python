```python
# 랜덤 포레스트
import pandas as pd
```

```python
# 붓꽃의 데이터 읽어 들이기
csv = pd.read_csv("../Data/iris.csv")
```

```python
csv.head()
```

```python
# 필요한 열 추출하기
csv_data = csv[['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']]
csv_label = csv['Name']
```

```python
# 학습 전용 데이터와 테스트 전용 데이터로 나누기

from sklearn.model_selection import train_test_split

train_data, test_data, train_label, test_label = train_test_split(
    csv_data,
    csv_label,
)
```

```python
# 데이터 학습 시키고 예측하기
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
clf = RandomForestClassifier()

clf.fit(train_data, train_label)
```

```python
# 정확도
clf.score(test_data, test_label)
```

```python
# Confusion Matrix *********

metrics.accuracy_score(test_label, clf.predict(test_data))
# metrics.classification_report(clf.predict(test_data, test_label))
```

```python
# 데이터 획득하기

import pandas as pd
import urllib.request as req

local = "../Data/mushroom.csv"
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data"
req.urlretrieve(url, local)
print('OK')
```

```python
# 데이터 읽기
mr = pd.read_csv("../Data/mushroom.csv", header=None)
mr.head()
```

# Data 정제하기

```python
 # 연습하기
 print(ord('a'))
 print(chr(97))
```

```python
# 데이터 내부의 문자를 숫자로 변환하기
label = []
data = []

for row_index, row in mr.iterrows():
    # print(row_index, row)
    # print(row.loc[1:])
    label.append(row.loc[0])
    row_data = []
    for v in row.loc[1:]:
        row_data.append(ord(v))
    data.append(row_data)

print(data[0:20])
```

```python
# label을 DataFrame으로 만들기
labelTemp = pd.DataFrame(label)
labelTemp.head()
```

```python
# data를 DataFrame으로 만들고 Column을 겹치지 않게 하기
dataTemp = pd.DataFrame(data)
dataTemp.rename(columns=lambda x: x+1, inplace=True)
dataTemp.head()
```

```python
# 2개의 DataFrame을 합치기
mr2 = pd.concat([labelTemp, dataTemp], axis='columns')
mr2.head()
```

```python
# RandomForest로 정확도 측정해 보기 

train_data, test_data, train_label, test_label = train_test_split(
    data,
    label,
)

clf.fit(train_data, train_label)
clf.score(test_data, test_label)
```

  이때 사용하는 것이 One Hot Encoding 이다.

```python
# 1번열 data의 종류 확인
mr2[1].unique()
```
